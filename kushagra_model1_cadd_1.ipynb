IMPORT PYTHON LIBRARIES LIKE PANDA as pd

```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

TARGET SEARCH FOR DAIBETES 2 PROTEIN (DIPEPTIDYL PEPTIDASE 4) namely DPP4

```python
# Target search for diabetes(type 2)
target_diabetes = new_client.target
target_query = target_diabetes.search('dipeptidyl peptidase 4')
targets = pd.DataFrame.from_dict(target_query)
#displaying results in a tabular format from chembl database
targets
```

SELECTED TARDET PROTEIN DPP4 IS SELECTED FOR HOMO SAPEINS

```python
final_target = targets.target_chembl_id[4]
final_target
```

Here, we will retrieve only bioactivity data for human dpp4  that are reported as pChEMBL values.

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=final_target).filter(standard_type="IC50")
df = pd.DataFrame.from_dict(res)
df
```

```python
df.to_csv('DPP4 raw data chembl 1.csv', index=False)
```

```python
#line1-remove data from table where ic50 is absent, line 2-remove rows where chemical structre is missing
df2 = df[df.standard_value.notna()]
df2 = df2[df.canonical_smiles.notna()]
df2
```

```python
#count number of molecules, with their uniqueness(no repeatation of compound)
len(df2.canonical_smiles.unique())
```

```python
#line1:- a buildin functioin in panda , it scan smiles column to avoid multiple entry of single molecule
df2_nr = df2.drop_duplicates(['canonical_smiles'])
df2_nr
```

Data pre-processing of the bioactivity data
Combine the 3 columns (molecule_chembl_id,canonical_smiles,standard_value) and bioactivity_class into a DataFrame

```python
#this line of code remove unnessecary data from the source
#line1:-Yeh simply ek list bana raha hai un sirf 3 columns ki, jo humare model ke liye actually zaroori hain.
#line2-pichle table (df2_nr) par ek filter lagata hai. Yeh computer ko bolta hai: "Mujhe sirf yeh 3 columns chahiye jo maine selection list mein likhe hain. Baaki ke saare 40+ columns delete kar do aur is naye, chote, aur saaf table ko df3 naam
selection = ['molecule_chembl_id','canonical_smiles','standard_value']
df3 = df2_nr[selection]
df3
```

Now we classify the bioactivity data into active , active and intermediate compounds

```python
df3.to_csv('DPP4 raw data chembl 1.csv', index=False)
```

```python
df4 = pd.read_csv('DPP4 raw data chembl 1.csv')

bioactivity_threshold = []
for i in df4.standard_value:
  if float(i) >= 10000:
    bioactivity_threshold.append("inactive")
  elif float(i) <= 1000:
    bioactivity_threshold.append("active")
  else:
    bioactivity_threshold.append("intermediate")


bioactivity_class = pd.Series(bioactivity_threshold, name='class')
df5 = pd.concat([df4, bioactivity_class], axis=1)
df5
```
